$\textbf{a)}$ Numerical resolution of the harmonic oscillator equation $\ddot{x}=-x$ using Euler's method.

```python
import numpy as np
import matplotlib.pyplot as plt


a=0
b=200
H=0.01
f=lambda x: -x
N=int((b-a)/H+1)
t=np.linspace(a,b,N)
x=np.zeros(len(t))
v=np.zeros(len(t))
x[0]=1
v[0]=-1
for j in range(1,N):
    x[j]=x[j-1]+v[j-1]*H
    v[j]=v[j-1]+f(x[j-1])*H
print("The solution is x= ",x)

print()
print()
plt.figure(figsize = (12, 8))
plt.plot(t,x)
plt.plot(t,np.cos(t)-np.sin(t))
plt.title("Approximate solution and exact solution")
plt.xlabel("time")
plt.ylabel("x(t)")
plt.grid()
plt.legend(loc='right')
plt.show()
```

$\textbf{b)}$ Numerical resolution of the harmonic oscillator equation $\ddot{x}=-x$ using Verlet method.

```python
import numpy as np
import matplotlib.pyplot as plt


a=0
b=200
H=0.01
f=lambda x: -x
N=int((b-a)/H+1)
t=np.linspace(a,b,N)
x=np.zeros(len(t))
x[0]=1
v0=-1
x[1]=x[0]+v0*H+(0.5)*f(x[0])*H
for j in range(1,N-1):
    x[j+1]=2*x[j]-x[j-1]+f(x[j])*H**2
print("The solution is x= ",x)

print()
print()
plt.figure(figsize = (12, 8))
plt.plot(t,x)
plt.plot(t,np.cos(t)-np.sin(t))
plt.title("Approximate solution and exact solution")
plt.xlabel("time")
plt.ylabel("x(t)")
plt.grid()
plt.legend(loc='right')
plt.show()
```

$\textbf{c)}$ Numerical resolution of the harmonic oscillator equation $\ddot{x}=-\sin x$ using Verlet method.

```python
import numpy as np
import matplotlib.pyplot as plt


a=0
b=200
H=0.01
f=lambda x: -np.sin(x)
N=int((b-a)/H+1)
t=np.linspace(a,b,N)
x=np.zeros(len(t))
v=np.zeros(len(t))
x[0]=1
v[0]=-1
x[1]=x[0]+v[0]*H+(0.5)*f(x[0])*H
for j in range(1,N-1):
    x[j+1]=2*x[j]-x[j-1]+f(x[j])*H**2
    v[j]=(1/(2*H))*(x[j+1]-x[j-1])
print("The solution is x= ",x)
print()
print("The speed v= ",v)

print()
print()
plt.figure(figsize = (12, 8))
plt.plot(t,x)
plt.title("Approximate solution x(t)")
plt.xlabel("time")
plt.ylabel("x(t)")
plt.grid()
plt.legend(loc='right')
plt.show()


plt.figure(figsize = (12, 8))
plt.plot(v,x)
plt.title("Approximate solution of x function of v")
plt.xlabel("v(t)")
plt.ylabel("x(t)")
plt.grid()
plt.legend(loc='right')
plt.show()
```

# $\beta =0.02$

```python
import numpy as np
import matplotlib.pyplot as plt

c_1=0.02
a=0
b=200
H=0.01
f=lambda x,v: -x-c_1*v
N=int((b-a)/H+1)
t=np.linspace(a,b,N)
x=np.zeros(len(t))
v=np.zeros(len(t))
x[0]=1
v[0]=-1
for j in range(0,N-1):
    x[j+1]=x[j]+H*v[j]+(0.5)*H**2*f(x[j],v[j])
    v[j+1]=v[j]+(0.5)*H*(f(x[j],v[j])+f(x[j+1],v[j+1]))
print("The solution is x= ",x)
print()
print("The speed v= ",v)

print()
print()
plt.figure(figsize = (12, 8))
plt.plot(t,x)
plt.title("Approximate solution x(t)")
plt.xlabel("time")
plt.ylabel("x(t)")
plt.grid()
plt.legend(loc='right')
plt.show()


plt.figure(figsize = (12, 8))
plt.plot(v,x)
plt.title("Approximate solution x(t) function of v(t)")
plt.xlabel("v(t)")
plt.ylabel("x(t)")
plt.grid()
plt.legend(loc='right')
plt.show()
```
